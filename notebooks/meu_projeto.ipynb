```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
```

```python
bits = 8
qc = QuantumCircuit(bits, bits)

for i in range(bits):
    qc.h(i)

qc.measure(range(bits), range(bits))

backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=1000, memory=True) 
result = job.result()

bitstring = result.get_memory()[0]
numero_decimal = int(bitstring, 2)

print(f"Resultado Binário: {bitstring}")
print(f"Número Inteiro Gerado: {numero_decimal}")

qc.draw(output='mpl')
```

```python
from qiskit.visualization import plot_histogram

# Pega a contagem de quantas vezes cada número apareceu
counts = result.get_counts()

# Exibe o gráfico de distribuição
plot_histogram(counts)
```
